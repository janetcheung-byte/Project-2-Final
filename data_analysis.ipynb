```python
from path import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import hvplot.pandas
import plotly.express as px
```

```python
# Loading data
file_path = Path("data/crops.csv")
crops_df = pd.read_csv(file_path)
crops_df.head()
```

```python
crops_df = crops_df.drop(columns=["state_alpha", "state_fips_code",'county_code','state_ansi'])
crops_df.tail()
```

```python
soybeans = crops_df.set_index(['county_name','state_name','year'])
soybeans.head()
```

```python
soybeans.sort_index(inplace=True)
soybeans.head()
```

```python
soybeans = soybeans.drop(columns=['AREA HARVESTED', 'total_acres','commodity_desc'])
soybeans
```

```python
total_events=soybeans.hvplot(kind='bar',x='year', y='total_events',groupby=['state_name','county_name'], color='purple', title='Total Loss Events')
total_events
```

```python
total_imdemnity=soybeans.hvplot(kind='bar',x='year', y='total_imdemnity', groupby=['state_name','county_name'], color='blue', title='Total Indemnity Amount').opts(yformatter="%.0f")
total_imdemnity
```
